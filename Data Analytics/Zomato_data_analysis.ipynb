```python
#Country-code.xlsx contains data about Country code with corresponding Countries
#reading data and storing into df dataframe
df_country=pd.read_excel('Country-code.xlsx')
df_country
```

```python
df_country.info() #getting information about data type and null values in Country code data
```

```python
#Merging Country-code data with Zomato data using merge
data=pd.merge(df,df_country,on='Country Code', how='left') # here how means where to add new column in initial dataframe
data #We can see Country column is added in left end of dataframe
```

```python
data.columns
```

```python
data['Country'].value_counts()
```

```python
country_names=data.Country.value_counts().index #storing country names or index in above value counts array
country_val=data.Country.value_counts().values ##storing country count or values in above value counts array
```

```python
# Pie Chart- Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
#grouping the data set on basis of Aggregate rating, rating color and rating text and reseting the index by adding rating count
#coulumn which stores value_counts of particular combo.
ratings=data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
#plotting pair plot using seaborn
sns.pairplot(ratings,hue='Rating text',height=6)
```

Insights from Aggregate rating vs Rating count

When Rating is between 4.5 to 4.9---> Excellent

When Rating are between 4.0 to 3.4--->very good

when Rating is between 3.5 to 3.9----> good

when Rating is between 3.0 to 3.4----> average

when Rating is between 2.5 to 2.9----> average

when Rating is between 2.0 to 2.4----> Poor 


```python
#Plotting bar plot to get insights about rating counts for aggregate ratings
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

From above barplot we concluded that

1.count of not rated or white color is greatest of all others

2.Maximum number of rating are between 2.8 to 3.9

```python
##plotting Count_plot to observe most rated Rating color
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','black'])
```

Here we observed that orange or average 3.0 to 3.4 is most rated category
